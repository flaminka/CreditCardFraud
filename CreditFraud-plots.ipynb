```python
import pandas as pd
#import numpy as np
#sciezka = "C:/Users/USER/Desktop"
#dane = pd.read_csv(sciezka+"creditcard.cs")
#dane=np.genfromtxt("C:/Users/USER/Desktop/vcreditcard.csv", unpack=True, skiprows = 1)
dane = pd.read_csv("C:/Users/USER/Desktop/creditcard.csv",encoding='latin1')

dane.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
import dateutil
su=sum(dane['Class']==1)
print ("In our data we have %d frouds." %su)

classes = pd.value_counts(dane['Class'], sort = True).sort_index()
classes.plot(kind = 'bar')
plt.title("Fraud class histogram")
plt.xlabel("Class")
plt.ylabel("Frequency")
plt.show()
```

```python
print ("The plot show how many frouds were in time." )

f, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10,10))


ax1.hist(dane.Time[dane.Class == 1], color='red')
ax1.set_title('Fraud')


ax2.hist(dane.Time[dane.Class == 0], color='green')
ax2.set_title('Normal')

plt.xlabel('Time (in Seconds)')
plt.ylabel('Number of Transactions')
plt.show()
```

```python
f, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10,6))

ax1.scatter(dane.Time[dane.Class == 1], dane.Amount[dane.Class == 1], color='red')
ax1.set_title('Fraud')

ax2.scatter(dane.Time[dane.Class == 0], dane.Amount[dane.Class == 0], color='green')
ax2.set_title('Normal')

plt.xlabel('Time (in Seconds)')
plt.ylabel('Amount')
plt.show()
```

```python
#from matplotlib import pyplot as plt
n0=plt.scatter(dane.Time[dane.Class == 0], dane.Amount[dane.Class == 0],color='green', label='Normal')
n1=plt.scatter(dane.Time[dane.Class == 1], dane.Amount[dane.Class == 1],color='red', label='Froud')
# Add a legend at the lower center
plt.legend((n0, n1),('Normal', 'Froud'),scatterpoints=1, loc='upper left', ncol=1, fontsize=8)

# Add axis labels and title
plt.xlabel('Time')
plt.ylabel('Amount')
plt.title('Amount vs time')
plt.rcParams['font.size']=10
plt.show()
```
